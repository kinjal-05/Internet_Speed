```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\KINJAL\\Desktop\\Data Science\\InternetSpeed\\Internet_Speed.csv')
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df['Connection_type_DSL'] = df['Connection_type_DSL'].astype('category')
df['Connection_type_Cable'] = df['Connection_type_Cable'].astype('category')
df['Connection_type_Fiber'] = df['Connection_type_Fiber'].astype('category')
```

```python
df
```

```python
continuous_vars = ['Ping_latency', 'Download_speed', 'Upload_speed', 'Packet_loss_rate', 'Router_distance', 'Signal_strength']

plt.figure(figsize=(15, 10))
for i, var in enumerate(continuous_vars):
    plt.subplot(2, 3, i+1)
    sns.histplot(df[var], bins=30, kde=True,color='g',edgecolor='r')
    plt.title(f'Distribution of {var}')
    plt.xlabel(var)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.countplot(data=df, x='Connection_type_DSL')
plt.title('Distribution of DSL Connections')

plt.subplot(1, 3, 2)
sns.countplot(data=df, x='Connection_type_Cable')
plt.title('Distribution of Cable Connections')

plt.subplot(1, 3, 3)
sns.countplot(data=df, x='Connection_type_Fiber')
plt.title('Distribution of Fiber Connections')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df['Connection_type'] = np.select(
    [df['Connection_type_DSL'] == 1, df['Connection_type_Cable'] == 1, df['Connection_type_Fiber'] == 1],
    ['DSL', 'Cable', 'Fiber'], default='Unknown'
)

sns.boxplot(data=df, x='Connection_type', y='Download_speed')
plt.title('Download Speed by Connection Type')
plt.xlabel('Connection Type')
plt.ylabel('Download Speed')
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=df, x='Network_congestion', y='Download_speed', hue='Connection_type')
plt.title('Network Congestion vs Download Speed')
plt.xlabel('Network Congestion')
plt.ylabel('Download Speed')

plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='Network_congestion', y='Upload_speed', hue='Connection_type')
plt.title('Network Congestion vs Upload Speed')
plt.xlabel('Network Congestion')
plt.ylabel('Upload Speed')

plt.tight_layout()
plt.show()
```
